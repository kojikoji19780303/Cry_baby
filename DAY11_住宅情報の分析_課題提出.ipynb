【問題5】用語の調査

これらを調査してまとめてください。
尖度
歪度

歪度とは、データの非対称性の度合いを示す度数です。方向は違えど、どちらも分布が「歪んでいる」ことが確認できます。統計学的な表現だと、「右の裾が長い」、「左の裾が長い」という言い方をします。こうした確率分布の「歪み」を表す指標が歪度です。


尖度は、分布の中心（期待値付近）に十分な確率が分布しており、かつ長い裾を引いている分布のことを「尖っている」という。尖度は正規分布と比べて、尖度が大きければ鋭いピークと長く太い裾を持った分布を持ち、 尖度が小さければより丸みがかったピークと短く細い尾を持った分布であるという事が判断できる。

【問題6】分布の確認

ポイント：対数変換をするとデータが正規分布に近づく傾向にあり、精度が上がる場合がある。seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

目的変数に対して対数変換を行う。

対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

```python
# 歪度 
df["SalePrice"].skew()
```

```python
# 尖度
df["SalePrice"].kurtosis()
```

```python
sns.distplot(df["SalePrice"], kde=False, rug=False, bins=100)

# https://www.codexa.net/seaborn-python/#:~:text=sns.distplot%20%E3%81%AE%E5%BC%95%E6%95%B0%E3%81%AB%E3%81%AF%E3%80%81Pandas%E3%81%AESeries%E3%80%81Numpy%E3%81%AE1d-array%E3%80%81%E3%82%82%E3%81%97%E3%81%8F%E3%81%AFPython%E3%81%AE%E3%83%AA%E3%82%B9%E3%83%88%E3%82%92%E6%8C%87%E5%AE%9A%E3%81%97%E3%81%BE%E3%81%99%E3%80%82,%E3%81%93%E3%81%AE%E3%82%B0%E3%83%A9%E3%83%95%E3%81%8B%E3%82%89%E3%80%8120%E4%BB%A3%E5%89%8D%E5%8D%8A%E3%81%AE%E4%BA%BA%E3%81%8C%E6%9C%80%E3%82%82%E5%A4%9A%E3%81%84%E3%81%93%E3%81%A8%E3%81%8C%E3%82%8F%E3%81%8B%E3%82%8A%E3%81%BE%E3%81%99%E3%81%AD%E3%80%82%20%E4%BB%8A%E5%9B%9E%E3%81%AE%E3%83%92%E3%82%B9%E3%83%88%E3%82%B0%E3%83%A9%E3%83%A0%E3%81%AF%E5%B9%854%E3%80%81%E5%85%A8%E9%83%A8%E3%81%A720%E3%81%AE%E9%9A%8E%E7%B4%9A%EF%BC%88bin%EF%BC%89%E3%82%92%E6%8C%81%E3%81%A3%E3%81%A6%E3%81%84%E3%81%BE%E3%81%99%E3%80%82
```

```python
# 対数変換
df_log = df["SalePrice"].apply(np.log)

# 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する
```

```python
# 歪度 
df_log.skew()
```

```python
# 尖度
df_log.kurtosis()
```

```python
sns.distplot(df_log, kde=False, rug=False, bins=100)
```

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。

元のデータでは尖度が6.53であったが、対数変換により尖度が0.81となった。
歪度も1.88から0.12になった。分布が正規分布に近くなった。

【問題7】相関係数の確認
特徴量に対してマークダウンで説明や考察を書いてください。


相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。

選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

```python
# 相関係数行列のヒートマップを作る
df_corr = df.corr()
colormap = plt.cm.RdBu
plt.subplots(figsize=(15, 12))
plt.title('Correlation matrix')
sns.heatmap(df_corr,linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=False)
plt.show()
```

```python
# SalePriceとの相関をソート
df_corr_saleprice  = df_corr["SalePrice"].sort_values(ascending=False)
print(df_corr_saleprice[1:11])
```

```python
df_high_corr_saleprice = df[df_corr_saleprice[1:11].index] # カラム名取ってきてる
display(df_high_corr_saleprice.head())

df_high_corr_saleprice_corr = df_high_corr_saleprice.corr() # 相関行列算出

colormap = plt.cm.RdBu
plt.subplots(figsize=(7, 7))
plt.title('Correlation matrix')
sns.heatmap(df_high_corr_saleprice_corr,linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=False)
plt.show()
```
